```python
#applying onehot encoder to encode the categorical data into numeric 

numeric_data1=pd.get_dummies(string_data)

numeric_data1.head(3)
```

```python
df_combined = pd.concat([numeric_data,numeric_data1], axis=1)

df_combined.head(3)
```

```python
plt.figure(figsize=(35,35))
sns.heatmap(df_combined.corr(),annot=True,fmt='.0%')
```

```python
# we can see with this heat map that attrition has good corelation with over time ,employees doing overtime are leaving 
# employess who are single are not sticking with the company as compare to married employees
#sales representative are also leaving the job
```
